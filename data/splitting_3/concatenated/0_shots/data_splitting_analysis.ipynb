```python
import pandas as pd
import numpy as np
import rdkit
from rdkit import Chem
```

# Data loading 

```python

raw_df3 = pd.read_csv('/scratch/yw5806/Multistep/data/splitting/rxn_0shot.csv')

raw_df3.describe()
raw_df3.info()
raw_df3.head()
```

```python


test_source = open('test.source', 'w')
test_target = open('test.target', 'w')
```

# Write input files

```python
for i, row in raw_df3.iterrows():
    if row["Method"] == "E":
        #print(i)
        step1_reactants = Chem.CanonSmiles( ".".join( [row['Reagent_A'], row['Reagent_B']] ) )
        step_1_reagents = Chem.CanonSmiles(row['step_1_reagents'])
        step2_reactants = Chem.CanonSmiles(row['Reagent_C'])
        step_2_reagents = Chem.CanonSmiles(row['step_2_reagents'])
        product = Chem.CanonSmiles(row['Product'])
        print(step1_reactants + "." + step_1_reagents + "//" + step2_reactants + "." + step_2_reagents, file=test_source)
        print(product, file=test_target)
    else:
        # for other methods, step2 is deprotection and there's no reactants involved.
        step1_reactants = Chem.CanonSmiles( ".".join( [row['Reagent_A'], row['Reagent_B']] ) )
        step_1_reagents = Chem.CanonSmiles(row['step_1_reagents'])

        step_2_reagents = Chem.CanonSmiles(row['step_2_reagents'])
        step3_reactants = Chem.CanonSmiles(row['Reagent_C'])
        step_3_reagents = Chem.CanonSmiles(row['step_3_reagents'])
        product = Chem.CanonSmiles(row['Product'])        
        print(step1_reactants + "." + step_1_reagents + "//" + step_2_reagents + "//" + step3_reactants + "." + step_3_reagents, file=test_source)
        print(product, file=test_target)
```

```python

test_source.close()
test_target.close()
```
