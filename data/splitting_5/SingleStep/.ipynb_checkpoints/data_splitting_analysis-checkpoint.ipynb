```python
import pandas as pd
import numpy as np
import rdkit
from rdkit import Chem
```

# Data loading

```python

raw_df1 = pd.read_csv('/scratch/yw5806/Multistep_final/data/splitting_5/rxn_train.csv')
raw_df2 = pd.read_csv('/scratch/yw5806/Multistep_final/data/splitting_5/rxn_vali.csv')
raw_df3 = pd.read_csv('/scratch/yw5806/Multistep_final/data/splitting_5/rxn_test.csv')

raw_df3.describe()
raw_df3.info()
raw_df3.head()
```

```python
train_source = open('train.source', 'w')
train_target = open('train.target', 'w')

val_source = open('val.source', 'w')
val_target = open('val.target', 'w')

test_source = open('test.source', 'w')
test_target = open('test.target', 'w')
```

# Write input files

```python
for i, row in raw_df1.iterrows():

    if row["Method"] == "E":
        #print(i)
        step1_reactants = Chem.CanonSmiles( ".".join( [row['Reagent_A'], row['Reagent_B']] ) )
        step_1_reagents = Chem.CanonSmiles(row['step_1_reagents'])
        step2_reactants = Chem.CanonSmiles(row['Reagent_C'])
        step_2_reagents = Chem.CanonSmiles(row['step_2_reagents'])
        interm_1 = Chem.CanonSmiles(row['step_1_intermediate'])
        product = Chem.CanonSmiles(row['Product'])

        print(step1_reactants + "." + step_1_reagents + ">>" , file=train_source)
        print(interm_1 + "."  + step2_reactants + "." + step_2_reagents  + ">>" , file=train_source)

        print(interm_1, file=train_target)
        print(product, file=train_target)
    else:
        # for other methods, step2 is deprotection and there's no reactants involved.
        step1_reactants = Chem.CanonSmiles( ".".join( [row['Reagent_A'], row['Reagent_B']] ) )
        step_1_reagents = Chem.CanonSmiles(row['step_1_reagents'])
        interm_1 = Chem.CanonSmiles(row['step_1_intermediate'])

        step_2_reagents = Chem.CanonSmiles(row['step_2_reagents'])
        interm_2 = Chem.CanonSmiles(row['step_2_intermediate'])
        step3_reactants = Chem.CanonSmiles(row['Reagent_C'])
        step_3_reagents = Chem.CanonSmiles(row['step_3_reagents'])
        product = Chem.CanonSmiles(row['Product'])
        print(step1_reactants + "." + step_1_reagents  + ">>" , file=train_source)
        print(interm_1 + "." + step_2_reagents  + ">>" , file=train_source)
        print(interm_2 + "." + step3_reactants + "." + step_3_reagents + ">>" , file=train_source)

        print(interm_1, file=train_target)
        print(interm_2, file=train_target)
        print(product, file=train_target)
```
